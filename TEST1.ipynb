```python
import numpy as np
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from time import sleep
```

```python
#s=Service(r'C:\Users\lenovo\Assignment-1\chromedriver-win64\chromedriver-win64.exe')
driver=webdriver.Chrome()
driver.get("https://uk.investing.com/commodities/us-sugar-no11-historical-data")
#driver.maximize_window()

sleep(10)
ad_close=driver.find_element(By.XPATH,"(//button[@type='button'])[28]")
ad_close.click()
print("ADclosed")


time_frame = driver.find_element(By.XPATH,"//div[contains(@class, 'historical-data-v2_selection-arrow')]")
time_frame.click()

weekly=driver.find_element(By.XPATH,"//div[contains(@class, 'historical-data-v2_menu-row__oRAlf')][2]").click()
sleep(5)
table=driver.find_element(By.CLASS_NAME,'freeze-column-w-1')

headers = [header.text for header in table.find_elements(By.XPATH, './/thead//th')]
rows = []
for row in table.find_elements(By.XPATH, './/tbody//tr'):
    cells = [cell.text for cell in row.find_elements(By.XPATH, './/td')]
    if cells:
        rows.append(cells)

driver.quit()
```

```python
df = pd.DataFrame(rows, columns=headers)
df
```

```python
clean=df
X=clean.drop(['High','Low','Vol.'],axis=1)
X
```

```python
X["product_name"]='Sugar'
product_cols=X.pop('Product_name')
X.insert(0,"Product_name",product_cols)
```

```python
X.pop('product_name')
```

```python
X.rename(columns={'Date':'Product_Date'},inplace=True)
```

```python
X
```

```python
X['Product_Date']=pd.to_datetime(X['Product_Date'],format='%d-%m-%Y')
X.head()
```

```python
X['DAY'] = X['Product_Date'].dt.day_name()
X.head()
```

```python
import requests
api_url = 'https://api.exchangerate-api.com/v4/latest/USD'
response = requests.get(api_url)
data = response.json()

X['Price'] = X['Price'].replace({'\$': ''}, regex=True).astype(float)

usd_to_inr_rate = data['rates']['INR']

X['Final Price (INR)'] = X['Price'] * usd_to_inr_rate
```

```python
output=X.drop(['Change %','Price','Open'],axis=1)
```
